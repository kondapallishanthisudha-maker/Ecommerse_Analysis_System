# PRODUCTS

```python
products.head()
```

```python
products.info()
```

```python
products.isnull().sum()
```

```python
products.duplicated().sum()
```

```python
products["product_category_name"] = products["product_category_name"].fillna("Unknown")
```

```python
numeric_columns = [
    "product_name_lenght",
    "product_description_lenght",
    "product_photos_qty",
    "product_weight_g",
    "product_length_cm",
    "product_height_cm",
    "product_width_cm"
]

for col in numeric_columns:
    products[col] = products[col].fillna(products[col].median())
```

```python
products.isnull().sum()
```

```python
products.to_csv("../dataset/products_clean.csv", index=False)
```

## Products Dataset Cleaning

- Checked dataset structure
- Identified missing values
- Filled missing product categories with "Unknown"
- Filled missing numeric values using the median
- Verified no missing values remain
- Saved cleaned dataset as `products_clean.csv`
